```python
'''
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
'''

# Alternative to iloc
x=df.drop('deposit',axis=1)
y=df['deposit']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#HARD MARGIN
linsvc=LinearSVC(random_state=0)  #Support vector machine linear
linsvc.fit(xtrain,ytrain)
ypred=linsvc.predict(xtest)
ypred
```

```python
from sklearn.metrics import confusion_matrix,classification_report #classification report
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
train=linsvc.score(xtrain,ytrain)
test=linsvc.score(xtest,ytest)
print(f"Training score : {train} \nTesting score:{test}")
```

```python
#SOFT MARGIN
linsvc1=LinearSVC(random_state=0,C=0.8)  #c is 0.8
linsvc1.fit(xtrain,ytrain)
ypred=linsvc1.predict(xtest)
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
train=linsvc1.score(xtrain,ytrain)
test=linsvc1.score(xtest,ytest)
print(f"Training score:{train}\nTesting score:{test}")
```

```python
# SVC with kernel- polynomial -Hard margin
poly=SVC(random_state=0,kernel='poly',degree=2)
poly.fit(xtrain,ytrain)
ypred=poly.predict(xtest)
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
train=poly.score(xtrain,ytrain)
test=poly.score(xtest,ytest)
print(f"Training score:{train}\nTesting score:{test}")
```

```python
# SVC with kernel- polynomial -soft margin
poly=SVC(random_state=0,kernel='poly',degree=2,C=0.5)
poly.fit(xtrain,ytrain)
ypred=poly.predict(xtest)
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
train=poly.score(xtrain,ytrain)
test=poly.score(xtest,ytest)
print(f"Training score:{train}\nTesting score:{test}")
```

```python
# SVC with kernel- rbf -soft margin
rbf=SVC(random_state=0,kernel='rbf',C=0.5)
rbf.fit(xtrain,ytrain)
ypred=rbf.predict(xtest)
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
train=rbf.score(xtrain,ytrain)
test=rbf.score(xtest,ytest)
print(f"Training score:{train}\nTesting score:{test}")
```

```python
# SVC with kernel- radial -hard margin
rbf=SVC(random_state=0,kernel='rbf')
rbf.fit(xtrain,ytrain)
ypred=rbf.predict(xtest)
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
train=rbf.score(xtrain,ytrain)
test=rbf.score(xtest,ytest)
print(f"Training score:{train}\nTesting score:{test}")
```
