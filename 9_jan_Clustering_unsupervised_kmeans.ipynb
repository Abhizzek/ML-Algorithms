```python
X=df.iloc[:,[3,4]].values
X
```

```python
#find the number of clusters
#Model building

import scipy.cluster.hierarchy as sch           #based on
sch.dendrogram(sch.linkage(X,method='ward'))
plt.xlabel('Customer')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
#PURPOSE BASED METHOD, euclidean distance method
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage='ward')
y_pred=model.fit_predict(X)
y_pred
```

```python
plt.scatter(X[y_pred == 0,0],X[y_pred == 0,1],s=100, c='red',label='Cluster 1')
plt.scatter(X[y_pred == 1,0],X[y_pred == 1,1],s=100, c='blue',label='Cluster 2')
plt.scatter(X[y_pred == 2,0],X[y_pred == 2,1],s=100, c='green',label='Cluster 3')
plt.scatter(X[y_pred == 3,0],X[y_pred == 3,1],s=100, c='magenta',label='Cluster 4')
plt.scatter(X[y_pred == 4,0],X[y_pred == 4,1],s=100, c='cyan',label='Cluster 5')
plt.legend()
plt.show()
```

**SAME DATASET , TO FIND CLUSTERS elbow method**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Customers.csv')
df
```

```python
X=df.iloc[:,[3,4]].values
X
```

```python
#ELBOW METHOD
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    model =KMeans(n_clusters=i,init='k-means++', random_state=42)
    model.fit(X)
    wcss.append(model.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model = KMeans(n_clusters=5,init='k-means++',random_state=42)
y_pred=model.fit_predict(X)
```

```python
plt.scatter(X[y_pred == 0,0],X[y_pred == 0,1],s=100, c='red',label='Cluster 1')
plt.scatter(X[y_pred == 1,0],X[y_pred == 1,1],s=100, c='blue',label='Cluster 2')
plt.scatter(X[y_pred == 2,0],X[y_pred == 2,1],s=100, c='green',label='Cluster 3')
plt.scatter(X[y_pred == 3,0],X[y_pred == 3,1],s=100, c='magenta',label='Cluster 4')
plt.scatter(X[y_pred == 4,0],X[y_pred == 4,1],s=100, c='cyan',label='Cluster 5')
plt.legend()
plt.show()
```
