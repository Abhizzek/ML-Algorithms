```python
# Replace missing values with the mean of each column  #alternative to imputation
df = df.fillna(df.mean())
df.head()
```

```python
df.isnull().sum()
```

```python
#renaming
df= df.rename(columns={'MEDV': 'Houseprice'})
df
```

```python
X=df.iloc[:,:-1].values #[:,0:13]  , rows & columns
y=df.iloc[:,-1].values  #[:, 13]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
#RELATIONSHIP BETWEEN THE 13 FEATURES
independent_feature_names = df.columns.difference(['Houseprice']) #to check dependence of 13 features, data.columns.difference(doesnt take common features)
lm_coff=pd.Series(lm.coef_,index=independent_feature_names)       #coff-slope
lm_coff
```

```python
lm_coff.plot(kind="bar")
```

```python
#Ridge Algorithm
from sklearn.linear_model import Ridge
rid=Ridge()
rid.fit(X_train,y_train)
```

```python
#RIDGE
independent_feature_names = df.columns.difference(['Houseprice'])
rid_coff=pd.Series(rid.coef_,index=independent_feature_names)
rid_coff
```

```python
rid_coff.plot(kind='bar')   #not alot of difference
```

```python
#LASSO
from sklearn.linear_model import Lasso
la=Lasso()
la.fit(X_train,y_train)
```

```python
independent_feature_names = df.columns.difference(['Houseprice'])
la_coff = pd.Series(la.coef_, index=independent_feature_names)
la_coff
```

```python
la_coff.plot(kind='bar')   #feature selection 6 out of 13
```

```python
from sklearn.linear_model import ElasticNet
elnet=ElasticNet()
elnet.fit(X_train, y_train)
```

```python
#ELASTIC NET   #maybe the features which is removed in LASSO  could be important
independent_feature_names = df.columns.difference(['Houseprice'])
elnet_coff=pd.Series(elnet.coef_,index=independent_feature_names)
elnet_coff
```

```python
elnet_coff.plot(kind='bar')  #only 4 features out of 13 wont be necessary
```

```python
#Prediction
lm_pred=lm.predict(X_test)  #linear model prediction
lm_pred
```

```python
#RIDGE PREDICTION
rid_pred=rid.predict(X_test)

rid_pred
```

```python
#LASSO PREDICTION
lam_pred=la.predict(X_test)
lam_pred
```

```python
#Elastic net
elnet_pred=elnet.predict(X_test)
elnet_pred
```

```python
#y_test - actual value, predicted value- between regularisations
#Ridge MSE, R2, Accuracy
from sklearn import metrics
metrics.mean_squared_error(y_test,rid_pred)
```

```python
metrics.r2_score(y_test,rid_pred)
```

```python
#Lasso MSE, R2, Accuracy
#from sklearn import metrics
metrics.mean_squared_error(y_test,lam_pred)
```

```python
la.score(X_test,y_test)
```

```python
metrics.mean_squared_error(y_test,elnet_pred)
```

```python
metrics.r2_score(y_test,elnet_pred)
```

```python
elnet.score(X_test,y_test)
```
